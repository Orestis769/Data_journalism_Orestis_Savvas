## 4) Πόσοι είναι οι άνδρες και πόσες οι γυναίκες? Ποια είναι η μέση διαφορά του πλούτου τους?

> **Συμβουλή:** Το δεύτερο μισό της ερώτησης χρειάζεται `groupby`, το πρώτο όχι.

```python
print(df.columns)
```

```python
# 4) Πόσοι είναι οι άνδρες και πόσες οι γυναίκες;

gender_counts = df["gender"].value_counts()

print(gender_counts)
```

```python
# 4) Μέσος πλούτος ανά φύλο

mean_wealth_by_gender = df.groupby("gender")["networthusbillion"].mean()

print(mean_wealth_by_gender)
```

```python
# Διαφορά μέσου πλούτου ανδρών - γυναικών

difference = (
    mean_wealth_by_gender["male"]
    - mean_wealth_by_gender["female"]
)

print("Μέση διαφορά πλούτου ανδρών - γυναικών:")
print(difference)
```

```python
# Πιο καθαρή εκτύπωση

print("Άνδρες:", gender_counts["male"])
print("Γυναίκες:", gender_counts["female"])

print("Μέσος πλούτος ανδρών:", mean_wealth_by_gender["male"])
print("Μέσος πλούτος γυναικών:", mean_wealth_by_gender["female"])
print("Διαφορά:", difference)
```

## 5) Ποιος είναι ο φτωχότερος δισεκατομμυριούχος? Ποιοι είναι οι 10 φτωχότεροι?

```python
# Βρίσκουμε τη γραμμή με τον μικρότερο πλούτο
poorest = df.loc[df["networthusbillion"].idxmin()]

print("Φτωχότερος δισεκατομμυριούχος:")
print(poorest[["name", "networthusbillion"]])
```
